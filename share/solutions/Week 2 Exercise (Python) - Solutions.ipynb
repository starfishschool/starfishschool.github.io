## Writing a Gaussian Function

In this cell, write the a gaussian function that takes the standard deviation as a parameter and a mean of zero. As a reminder, the formula for such a Gaussian is:

$g(x) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\Big(-\frac{1}{2}} \frac{x^2}{\sigma^2} \Big) $

```python
def gaussian_function(x, sigma):
    """
    This function creates a gaussian with a mean of zero with an arbitrary standard deviation
    """

    tmp1 = 1.0 / (sigma * np.sqrt(2 * np.pi))
    tmp2 = np.exp(-0.5 * (x ** 2) / (sigma ** 2))

    return tmp1 * tmp2
```

## Adding a Gaussian Scatter

Using a random variable function from NumPy, add a 20% Gaussian Scatter to your points from the data file

```python
noise = 0.2 * data * np.random.randn(*data.shape)

# Hot Tip: if you have a tuple of list that contains all the parameters you want to pass to a command
#          you can use the asterisk character (*) to expand it in the function instantiation

data_with_noise = data + noise
```

## Plotting
